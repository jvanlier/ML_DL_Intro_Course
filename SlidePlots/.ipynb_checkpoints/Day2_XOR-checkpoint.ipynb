```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams["font.size"] = 16
matplotlib.rcParams["figure.dpi"] = 125
```

```python
n = 100
vmax = 5

c1x1 = np.concatenate((
    np.random.uniform(low=-vmax, high=0, size=n),
    np.random.uniform(low=0, high=vmax, size=n)
))

c1x2 = np.concatenate((
    np.random.uniform(low=-vmax, high=0, size=n),
    np.random.uniform(low=0, high=vmax, size=n)
))

c2x1 = np.concatenate((
    np.random.uniform(low=0, high=vmax, size=n),
    np.random.uniform(low=-vmax, high=0, size=n)
))

c2x2 = np.concatenate((
    np.random.uniform(low=-vmax, high=0, size=n),
    np.random.uniform(low=0, high=vmax, size=n)
))
```

```python
plt.figure(figsize=(5, 6))
plt.scatter(c1x1, c1x2)
plt.scatter(c2x1, c2x2)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
plt.figure(figsize=(5, 6))
plt.scatter(c1x1 * c1x2, c1x2)
plt.scatter(c2x1 * c2x2, c2x2)
plt.xlabel("$x_3 = x_1x_2$")
plt.ylabel("$x_2$")
```
