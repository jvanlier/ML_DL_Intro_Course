```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
matplotlib.rcParams["font.size"] = 16
matplotlib.rcParams["figure.dpi"] = 125
```

```python
def f(x):
    return x * 2 + 100.000 + np.random.randn(len(x)) * 30
```

```python
XMIN = 50
XMAX = 150
xs = np.linspace(XMIN, XMAX, 30)
ys = f(xs)
data = pd.DataFrame({"x": xs, "y": ys})

# sns.lmplot("x", "y", data=data, ci=None, height=8)
def house_scatter():
    plt.figure(figsize=(6, 8))
    plt.scatter(data["x"], data["y"])
    plt.ylabel("house price (thousands of euros)")
    plt.xlabel("square meters")
    plt.grid()
    plt.ylim(0, )
    plt.xlim(XMIN, XMAX)
house_scatter()
```

```python
house_scatter()
xs_line = np.array((XMIN, XMAX))
plt.plot(xs_line, f(xs_line), color='C1')
```

```python
x1 = xs

def f2(x1, x2):
    return x1 * 2 + 100.000 + x2 * 20

x2 = []
for i in range(2, 8, 1):
    x2 += [i] * 5
x2 = np.array([n + random.randint(0, 2) for n in x2])
 
ys = f2(x1, x2)
```

```python
fig = plt.figure(figsize=(10, 10), dpi=200)
ax = fig.gca(projection='3d')

#ax.plot_surface(x_sm, y_sr, f2(x_sm, y_sr), alpha=.2, color='None', cmap='viridis')

x1_jit = x1 + np.random.randn(len(x1)) * 5
x2_jit = x2 #+ np.random.randn(len(x2)) * .5

ax.scatter(x1_jit, x2_jit, ys)

#ax.plot(x1, n_rooms, ys)

ax.set_xlabel("\nSquare meters")
ax.set_ylabel("\nNumber of rooms")
ax.set_zlabel("\nhouse price (1000s)", linespacing=3.1)

x_sm = np.arange(np.min(x1_jit), np.max(x1_jit) + 10, 1)
y_nr = np.arange(np.min(x2_jit), np.max(x2_jit) + 1, 1)
x_sm, y_sr = np.meshgrid(x_sm, y_nr)

ax.plot_surface(x_sm, y_sr, f2(x_sm, y_sr), color='None', cmap='Wistia', alpha=.5)
```

```python
def f(x):
    return (x - 20) ** 2 + 50

xs = np.arange(-10, 50 + 1, 1)
ys = f(xs)

plt.figure(dpi=200)

plt.plot(xs, ys)
plt.grid()
plt.ylim(0, )

plt.xlabel("Value of parameter $\\theta_0$")
plt.ylabel('"Cost" of solution')

```
