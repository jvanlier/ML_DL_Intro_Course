```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.signal import convolve2d
```

```python
imgo = Image.open("schiphol.jpeg")
img = np.array(imgo)
img = img[800:1250, 720:1200]
plt.figure(figsize=(10,10))
plt.imshow(img)
```

```python
imgo = Image.open("schiphol.jpeg").convert("L")
img = np.array(imgo)
img = img[800:1250, 720:1200]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
```

```python
vert = np.array(
    [[1, 0, -1],
     [1, 0, -1],
     [1, 0, -1]])

hori = vert.T
```

```python
all_edges = np.array(
    [[-1, -1, -1],
     [-1, 8, -1],
     [-1, -1, -1]])
```

```python
img_conv = convolve2d(img, hori, mode='valid')
```

```python
fig, (l, r) = plt.subplots(figsize=(20, 10), nrows=1, ncols=2, sharey=True)
l.imshow(img, cmap='gray')
r.imshow(img_conv, cmap='gray')
plt.tight_layout()
```

```python
img_conv = convolve2d(img, vert, mode='valid')

fig, (l, r) = plt.subplots(figsize=(20, 10), nrows=1, ncols=2, sharey=True)
l.imshow(img, cmap='gray')
r.imshow(img_conv, cmap='gray')
plt.tight_layout()
```
