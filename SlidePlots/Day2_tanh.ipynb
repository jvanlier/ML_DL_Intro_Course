```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
matplotlib.rcParams["font.size"] = 16
matplotlib.rcParams["figure.dpi"] = 125
```

```python
xs = np.arange(-10, 10, .1)
ys = np.tanh(xs)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(xs, ys)
plt.grid()
plt.ylabel("$\\tanh(z)$")
plt.xlabel("$z$")
plt.xlim(-3, 3)
plt.ylim(-1, 1)
```
