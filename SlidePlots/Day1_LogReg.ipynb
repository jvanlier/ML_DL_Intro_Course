```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
matplotlib.rcParams["font.size"] = 16
matplotlib.rcParams["figure.dpi"] = 125
```

```python
def logit(z):
    return 1 / (1 + np.exp(-z))
```

```python
xs = np.arange(-10, 10, .1)
ys = logit(xs)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(xs, ys)
plt.grid()
plt.ylabel("$g(z)$")
plt.xlabel("$z$")
plt.xlim(-5, 5)
plt.ylim(0, 1)
```
